```python
with open('truyenkieu.txt', 'r', encoding='utf-8') as f:
  data = f.read()

# import re
# data = re.sub('([.,!?():-])', r' \1 ', data)
# with open('truyenkieunew.txt', 'w') as f:
#   f.write(data)

vocab = sorted(list(set([w.strip() for w in data.split()])))
with open('vocab.txt', 'w') as f:
  for w in vocab:
    f.write(w + '\n')
print('vocab size:', len(vocab))
print('some words in our vocab:', vocab)
```
